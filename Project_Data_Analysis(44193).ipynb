### USA data

```python
DC =df[df['Country']=='United States']
DC
```

### USA data with State/Province as Index

```python
DC1 = DC.set_index('State/Province')
DC1
```

### Required Data for Comparison With South Africa Data

```python
DC2= DC1.iloc[:,2:7]
DC2
#Rated Power(kW),Discharge Duration at Rated Power(hrs),Storage Capacity(kWh) data and Status of different Projects or Plants.
```

### Combined data of South Africa and USA

```python
combined_df = pd.concat([dc,DC], keys=['South Africa','USA'])
combined_df
```

### Comparison of 'Sum of Rated Power(kW)' in South Africa and USA

```python
combined_df.loc['South Africa']['Rated Power (kW)'].sum()
# Sum of the Rated power(kW) of all the plants in South Africa.
```

```python
combined_df.loc['USA']['Rated Power (kW)'].sum()
# Sum of the Rated power(kW) of all the plants in USA.
```

```python
combined_df_country1 = combined_df.groupby(['Country'])['Rated Power (kW)'].sum()
combined_df_country1
# Combining the Sum of Rated power(kW) in South Africa and USA.
```

```python
combined_df_country1.plot.bar(x ='Country',y =' Rated Power (kW)', color= 'brown',rot=90);
plt.legend()
plt.title('Sum of Rated Power')
#Plotting the Sum of Rated power(kW) in South Africa and USA.
```

### Comparison of 'Maximum Rated Power (kW)' in South Africa and USA

```python
combined_df_country2 = combined_df.groupby(['Country'])['Rated Power (kW)'].max()
combined_df_country2
# Combining the maximum of Rated power(kW) in South Africa and USA.
```

```python
combined_df_country2.plot.bar(x ='Country',y =' Rated Power (kW)', color= 'purple',rot=90);
plt.legend()
plt.title('Max. Rated Power')
# Plotting the maximum of Rated power(kW) in South Africa and USA.
```

### Comparison of 'Maximum Storage Capacity(kWh) in South Africa and USA

```python
combined_df.loc['South Africa']['Storage Capacity (kWh)'].max()
# maximum Storage Capacity(kWh) in South Africa.
```

```python
combined_df.loc['USA']['Storage Capacity (kWh)'].max()
# maximum Storage Capacity(kWh) in USA.
```

```python
combined_df_country3 = combined_df.groupby(['Country'])['Storage Capacity (kWh)'].max()
combined_df_country3
# Combining the maximum Storage Capacity(kWh)in South Africa and USA.
```

```python
combined_df_country3.plot.bar(x ='Country',y ='Storage Capacity (kWh) ', color='orange',rot=90);
plt.legend()
plt.title('Storage Capacity ')
# Plotting the maximum Storage Capacity(kWh)in South Africa and USA.
```

### Comparison of ' Maximum Discharge Duration at Rated Power (hrs)' in South Africa and USA

```python
combined_df.loc['South Africa']['Discharge Duration at Rated Power (hrs)'].max()
# maximum Discharge Duration at Rated Power(hrs) in South Africa.
```

```python
combined_df.loc['USA']['Discharge Duration at Rated Power (hrs)'].max()
# maximum Discharge Duration at Rated Power(hrs) in USA.
```

```python
combined_df_country4 = combined_df.groupby(['Country'])['Discharge Duration at Rated Power (hrs)'].max()
combined_df_country4
# Combining the maximum Discharge Duration at Rated Power(hrs)in South Africa and USA.
```

```python
combined_df_country4.plot.bar(x ='Country',y ='Discharge Duration at Rated Power (hrs)', color='green',rot=90);
plt.legend(loc=2)
plt.title('Discharge Duration at Rated Power ')
# Plotting the maximum Discharge Duration at Rated Power(hrs)in South Africa and USA.
```

### Comparison between South Africa and USA

```python
plt.subplots(figsize=(12,10),dpi=200)
plt.subplots_adjust(hspace=0.3, wspace=0.3)
plt.subplot(2,2,1)
combined_df_country1.plot.bar(x ='Country',y =' Rated Power (kW)', color= 'brown',rot=0);
plt.legend()
plt.title('Sum of Rated Power')
plt.subplot(2,2,2)
combined_df_country2.plot.bar(x ='Country',y =' Rated Power (kW)', color= 'purple',rot=0);
plt.legend()
plt.title('Max. Rated Power')
plt.subplot(2,2,3)
combined_df_country3.plot.bar(x ='Country',y ='Storage Capacity (kWh) ', color='orange',rot=0);
plt.legend()
plt.title('Storage Capacity ')
plt.subplot(2,2,4)
combined_df_country4.plot.bar(x ='Country',y ='Discharge Duration at Rated Power (hrs)', color='green',rot=0);
plt.legend(loc=2)
plt.title('Discharge Duration at Rated Power ')
plt.show()
```
